# Pandas

### Pandas is an open-source software library for the Python programming language providing high-performance, easy-to-use data structures and analysis tools. It is a cornerstone of modern data science workflows, enabling efficient manipulation and analysis of structured data such as tables or time series.

### Handles data (tables like Excel)

```python
# Create DataFrame and View data

import pandas as pd

data = {
    "rooms": [2, 1, 3, 5, 6, 7],
    "distance": [1.5, 5, 1, 3, 2, 4.5],
    "wifi": [1, 0, 1, 0, 1, 0],
    "rating": [4.5, 3.0, 4.8, 3.2, 4.1, 4.4]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Shows the first 5 rows of your data

df.head() 
```

```python
# Shows the given first n number of rows of your data

df.head(2) 
```

```python
# Gives summary statistics of your data

df.describe()
```

```python
# Select column

df["rooms"]
```

```python
# Filter data

df[df["rating"] > 4]
```
